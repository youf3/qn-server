```python
import json  
import requests
import logging as log
import networkx as nx
from pyvis import network as net
from networkx import node_link_graph

from IPython.display import IFrame

def main():
    try:
        res = requests.get("http://localhost:8081/api/topology")
    except Exception as e:
        log.error(f"Could not retrieve Controller topology: {e}")
        return
    data = res.json()
    print (data)
    #show_topology(data[0])
    return data[0]
        
data = main()
```

```python
#print(data)

edge_style_map = {'quantum': {'color': 'blue', 'dashes': False},
                 'classic_clk': {'color': 'purple', 'dashes': True},
                 'classic_photon_gen': {'color': 'green', 'dashes': True},
                 'classic_bsm_result': {'color': 'brown', 'dashes': True},
                 'stabilization': {'color': 'teal'}}

node_style_map = {'QNode': 'orange',
                  'BSMNode': 'lightgreen',
                  'MNode': 'grey',
                  'OpticalSwitch': 'lightblue'}

nxg = node_link_graph(data, edges="edges")
g = net.Network(notebook=True, directed=True, cdn_resources='in_line')
g.from_nx(nxg)
for node in g.nodes:
    node['physics'] = False
    node['color'] = node_style_map.get(node['type'])
for edge in g.edges:
    edge.update(edge_style_map.get(edge['title']))
g.show_buttons(filter_ = [])
g.set_edge_smooth('dynamic')
#g.toggle_physics(False)
g.show("example.html")
```
